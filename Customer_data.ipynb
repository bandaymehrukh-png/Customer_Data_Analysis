```python

pt_map = {
    'Belt': 'Bags & Accessories',
    'Blouse': 'Apparel',
    'Boots': 'Footwear',
    'Coat': 'Outerwear',
    'Dress': 'Apparel',
    'Handbag': 'Bags & Accessories',
    'Hat': 'Bags & Accessories',
    'Jacket': 'Outerwear',
    'Jeans': 'Apparel',
    'Sandals': 'Footwear',
    'Scarf': 'Apparel',
    'Shirt': 'Apparel',
    'Shoes': 'Footwear',
    'Shorts': 'Apparel',
    'Skirt': 'Apparel',
    'Sunglasses': 'Luxury Goods',
    'Sweater': 'Apparel',
    'Watch': 'Luxury Goods'
}


df['Product Type'] = df['Product Type'].fillna(df['Item Purchased'].map(pt_map))


print(df['Product Type'].isnull().sum())
```

```python
df["Product Type"].isnull().sum()
```

```python
df["Review Rating"] = df.groupby("Product Type")["Review Rating"].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
df["Review Rating"].isnull().sum()
```

```python
df.columns=df.columns.str.lower()
```

```python
df.columns=df.columns.str.replace(" ", "_")
```

```python
df.columns
```

```python
#Feature Engineering
```

```python
#creating column Age_Group
labels=["young", "adult", "middle_aged", "senior"]
df['age_group']=pd.qcut(df["age"],q=4, labels=labels)
df[["age","age_group"]].head(10)
```

```python
#creating column purchase freq days
unique_values = df['frequency_of_purchases'].unique()
print(unique_values)
```

```python
frequency_mapping={'Fortnightly':14, 'Weekly':7, 'Annually':365, 'Quarterly':90, 'Bi-Weekly':14, 'Monthly':30,
 'Every 3 Months':120}
```

```python
df["purchase_freq_days"]=df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[["purchase_freq_days","frequency_of_purchases"]].head()
```

```python
df.to_csv('cleaned_customer_shopping_data.csv', index=False)
```
